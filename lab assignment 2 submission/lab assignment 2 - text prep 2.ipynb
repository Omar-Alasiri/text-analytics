```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Reviews.csv')
df.head()
```

```python
def label_sentiment(rating):
    if rating in [1, 2]:
        return 'negative'
    elif rating == 3:
        return 'neutral'
    elif rating in [4, 5]:
        return 'positive'
    else:
        return 'unknown'

df['sentiment'] = df['Score'].apply(label_sentiment)
result_df = df[['Text', 'Score', 'sentiment']]
result_df
```

```python
result_df.to_csv('labelled_reviews.csv', index=False)
print('Sentiment labels added and saved to labelled_reviews.csv')
```
